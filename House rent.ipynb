```python
mode_value=hp2['balcony'].mode()[0]
hp2['balcony'].fillna(mode_value,inplace=True)
mode_value2=hp2['bath'].mode()[0]
hp2['bath'].fillna(mode_value2,inplace=True)
hp3=hp2.dropna(subset=['size','location'])
hp3.isnull().sum()
```

```python
counts=hp3['size'].value_counts()
counts
```

```python
ranked=counts.rank(method='dense',ascending=False).astype(int)
ranked_values=pd.DataFrame({'Value':counts.index,'Rank':ranked})
ranked_values
```

```python
hp3['BHK']=hp3['size'].map(ranked_values.set_index('Value')['Rank'])
hp3.head()
```

```python
import re
def convert_to_float(value):
    match=re.findall(r"[\d.]+",value)
    if match:
        return float(match[0])
    return None
hp3['total_sqft']=hp3['total_sqft'].apply(convert_to_float)
```

## MODEL TESTING
### 1. LINEAR REGRESSION

```python
from sklearn.model_selection import train_test_split

X=hp6[['total_sqft','balcony','bath']]
y=hp6['price']

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=55)
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,root_mean_squared_error,r2_score

mse=mean_squared_error(y_test,y_pred)
rmse=root_mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)

print(f'Mean squared error: {mse}')
print(f'Root Mean Squared Error: {rmse}')
print(f'r2 Score: {r2}')
```

## 2.XGBOOST

```python
from sklearn.model_selection import RandomizedSearchCV,GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from scipy.stats import randint
```

```python

X=hp6[['total_sqft','balcony','bath','location','area_type']]
y=hp6['price']
X_encoded=pd.get_dummies(X,columns=['location','area_type'],drop_first=True)
X_train,X_test,y_train,y_test=train_test_split(X_encoded,y,test_size=0.2,random_state=55)
```

```python
param_dist={
    'n_estimators':randint(50,100),
    'max_features':['sqrt','log2',None],
    'max_depth':randint(1,20),
    'min_samples_split':randint(2,10),
    'min_samples_leaf':randint(1,10),
    }

rf_model=RandomForestRegressor()
random_search=RandomizedSearchCV(rf_model,param_distributions=param_dist,n_iter=100,scoring='neg_mean_squared_error',cv=3,random_state=42)
```

```python
random_search.fit(X_train,y_train)
```

```python
print(RandomForestRegressor().get_params().keys())
```

```python
print(f'best parameters for random forest: {random_search.best_params_}')
print(f'best r2 score for random forest: {random_search.best_score_}')
```

```python
xgb_param_grid={
    'n_estimators':[100,200,300],
    'max_depth':[3,5,7],
    'learning_rate':[0.01,0.1,0.2],
}

xgb_model=XGBRegressor()

xgb_grid_search=GridSearchCV(estimator=xgb_model,param_grid=xgb_param_grid,scoring='r2',cv=5,verbose=1,n_jobs=1)
```

```python
xgb_grid_search.fit(X_train,y_train)
```

```python
print(f'best parameters for XGB: {xgb_grid_search.best_params_}')
print(f'best r2 score for XGB: {xgb_grid_search.best_score_}')
```
